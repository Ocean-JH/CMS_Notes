### 1. [ÊñêÊ≥¢ÈÇ£Â•ëÊï∞Âàó](https://en.wikipedia.org/wiki/Fibonacci_sequence "ÊñêÊ≥¢ÈÇ£Â•ëÊï∞ÂàóÁî±0Âíå1ÂºÄÂßãÔºå‰πãÂêéÁöÑÊñêÊ≥¢ÈÇ£Â•ëÊï∞ÊòØÁî±‰πãÂâçÁöÑ‰∏§Êï∞Áõ∏Âä†ËÄåÂæóÂá∫„ÄÇ")

````
def fibonacci(n):
    if n==1 or n==2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
````

````
def rev_strs(strs):
    strs.reverse()
    reversed_strs = []
    for string in strs:
        rev_str = string[::-1]
        reversed_strs.append(rev_str)
    return reversed_strs
````

````
strs = ['we', 'love', 'python']
rev_strs(strs)
````

### 3. Âõ†ÂºèÂàÜËß£

````
def factorization(num):
    factor = []
    while num > 1:
        for i in range(num - 1):
            k = i + 2
            if num % k == 0:
                factor.append(k)
                num = int(num / k)
                break

    return factor
````

````
factorization(2001)
````

### 2. ÁêÜÊÉ≥Ê∞î‰ΩìÁä∂ÊÄÅÊñπÁ®ãÂèØËßÜÂåñ

````
import matplotlib.pyplot as plt
import numpy as np

# ÂÆö‰πâÂ∏∏Èáè
R = 0.08206  # L¬∑atm/mol¬∑K
n = 1.00  # mol
T1 = 298  # K
T2 = 398  # K
T3 = 498  # K

# ‰ΩìÁßØÈááÊ†∑
V = np.linspace(1, 20, 100)  # ÁîüÊàê‰ΩìÁßØÁöÑÁ≠âÈó¥Ë∑ùÂèñÊ†∑ÁÇπ

# ‰ΩøÁî®ÁêÜÊÉ≥Ê∞î‰ΩìÁä∂ÊÄÅÊñπÁ®ãPV = nRTËÆ°ÁÆóÂéãÂäõ
P1 = n * R * T1 / V  # 298K‰∏ãÁöÑÂéãÂäõ
P2 = n * R * T2 / V  # 398K‰∏ãÁöÑÂéãÂäõ
P3 = n * R * T3 / V  # 498K‰∏ãÁöÑÂéãÂäõ

# ÁªòÂà∂ÂõæË°®
plt.figure(figsize=(10, 6))
plt.plot(V, P1, 'b-', linewidth=2, label='298 K')  # ÁªòÂà∂298K‰∏ãÁöÑÊõ≤Á∫ø
plt.plot(V, P2, 'r-', linewidth=2, label='398 K')  # ÁªòÂà∂398K‰∏ãÁöÑÊõ≤Á∫ø
plt.plot(V, P3, 'g-', linewidth=2, label='498 K')  # ÁªòÂà∂498K‰∏ãÁöÑÊõ≤Á∫ø
plt.title('Pressure vs. Volume for 1.00 mol of helium gas at different temperatures', fontsize=16)
plt.xlabel('Volume (L)', fontsize=14)
plt.ylabel('Pressure (atm)', fontsize=14)
plt.grid(True)
plt.legend(fontsize=12)
plt.tick_params(axis='both', labelsize=12)
plt.show()
````

Ê≥®Ôºö‰ª•‰∏ã‰ª£Á†ÅÂèÇËÄÉ‰∫Ü[**ÂåñÂ≠¶ÂèçÂ∫îÂè∞Èò∂ÂõæÂø´ÈÄüÁªòÂà∂**](https://nb.bohrium.dp.tech/detail/4934994790)‰∏é[**PythonËÑöÊú¨ÊèêÂèñÊï∞ÊçÆÁªòÂà∂ÁîµÂåñÂ≠¶Âè∞Èò∂Âõæ**](https://jonyafei.github.io/2021/06/20/python%E8%84%9A%E6%9C%AC%E6%8F%90%E5%8F%96%E6%95%B0%E6%8D%AE%E7%BB%98%E5%88%B6%E7%94%B5%E5%8C%96%E5%AD%A6%E5%8F%B0%E9%98%B6%E5%9B%BE/)

````
import copy

import numpy as np
import matplotlib.pyplot as plt

from cycler import cycler
from dataclasses import dataclass
from typing import List, Optional
from pathlib import Path
````

````
@dataclass
class StepsGraph:
    data: list
    legend_txt: str = None
    labels: List[str] = None

    def __post_init__(self):
        if self.labels:
            if len(self.data) != len(self.labels):
                raise ValueError("The lengths of data and labels must be equal.")

    def plot_line_dot(
            self,
            ax: plt.Axes,
            color: Optional[str] = None,
            path_label: str = None,
            energy_label: bool = False, energy_label_fontsize=18,
            step_label: List[str] = None, step_label_fontsize=12,
    ):
        y = copy.deepcopy(self.data)
        y_max, y_min = np.max(y), np.min(y)  # Ëé∑ÂèñyÂÄºÁöÑÊúÄÂ§ßÂÄºÂíåÊúÄÂ∞èÂÄº
        y_bias = (y_max - y_min) / 25  # Ëé∑ÂèñÊñáÊú¨Ê†áÁ≠æyÊñπÂêëÂÅèÁßªÈáè
        x = range(len(y))
        if path_label is None:
            path_label = self.legend_txt
        if step_label is None:
            step_label = self.labels
        y_new = []
        x_new = []
        # 1.ÁîüÊàêÊñ∞ÁöÑXYÂùêÊ†áÁÇπÔºå‰∏™Êï∞Âä†ÂÄç
        for j in range(len(y)):
            y_new.append(y[j])
            y_new.append(y[j])
            x_new.append(2 * j)
            x_new.append(2 * j + 1)
        # 2.ÁªòÂà∂ËôöÁ∫øÊäòÁ∫øÂõæ
        if color is None:
            color = next(ax._get_lines.prop_cycler)['color']
        ax.plot(x_new, y_new, linestyle='--', linewidth=2, color=color,
                )

        # 3.ÁªòÂà∂ÂÆûÁ∫øÊäòÁ∫øÂõæ
        k = 0
        while k < len(y_new):
            x_line = [x_new[k], x_new[k + 1]]
            y_line = [y_new[k], y_new[k + 1]]
            ax.plot(x_line, y_line, linestyle='-', linewidth=4, color=color, label=path_label if k == 0 else None)
            k += 2

        # 4.Ê†áËÆ∞ËÉΩÈáèÂÄºÔºåÂÅèÁßªÈáèËßÜÂÖ∑‰ΩìÊÉÖÂÜµËÄåÂÆö
        if energy_label:
            for j in range(len(x)):
                ax.text(x[j] * 2 + 0.5, y[j] + y_bias, "{:.2f}".format(y[j]), fontsize=energy_label_fontsize,
                        color=color, horizontalalignment='center')

        now_minx, now_maxx = ax.get_xlim()
        now_miny, now_maxy = ax.get_ylim()
        if step_label:
            for j in range(len(x)):
                ax.text(
                    (x[j] * 2 + 0.5 - now_minx) / (now_maxx - now_minx),
                    0,
                    step_label[j],
                    fontsize=step_label_fontsize,
                    color=color,
                    horizontalalignment='center',
                    verticalalignment='top',
                    transform=ax.transAxes
                )
        ax.set_xticks([])
        ax.set_xticklabels([])
````

```python
def linear_conjugate_gradient(A, b, x0, tol=1e-5, max_iter):
    # Initializations
    x = x0; r= b - A @ x ; p = r 
    rsold = r.T @ r
    for i in range(max_iter):
        Ap = A @ p
        alpha = rsold / (p.T @ Ap)
        x += alpha * p                # Equation.(3)
        r -= alpha * Ap
        rsnew = r.T @ r
        # Check for convergence
        if np.sqrt(rsnew) < tol:
            break
        p = r + (rsnew / rsold) * p    # Equation.(2)
        rsold = rsnew
    return x
```

```python
    if self.previous_x is None:
```
